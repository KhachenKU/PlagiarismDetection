```python
len(classified_text)
```

```python
clean_token_eng = []
```

```python
for i in tqdm(classified_text):
    clean = []
    NE_text = ''
    NE = ''
    trigger = 0
    for j in i:
        text = j[0]
        text = lemmatizer.lemmatize(text.lower())
        if (text == '' or text == ' '):
            continue

        if (j[1] == 'O' or j[1] == 'PERCENT'):
            if trigger == 0:
                NE_text = text
                NE = 'O'
                clean.append((NE_text, NE))
            else:
                clean.append((NE_text, NE))
                NE_text = text
                NE = 'O'
                clean.append((NE_text, NE))
                trigger = 0
        elif j[1] == 'LOCATION':
            if trigger == 0:
                NE_text = text
                NE = 'LOCATION'
                trigger = 1
            else:
                if j[1] == NE:
                    NE_text += " " + text
                else:
                    clean.append((NE_text, NE))
                    NE_text = text
                    NE = j[1]
        elif j[1] == 'PERSON':
            if trigger == 0:
                NE_text = text
                NE= 'PERSON'
                trigger = 1
            else:
                if j[1] == NE:
                    NE_text += " " + text
                else:
                    clean.append((NE_text, NE))
                    NE_text = text
                    NE = j[1]
        elif j[1] == 'ORGANIZATION':
            if trigger == 0:
                NE_text = text
                NE = 'ORGANIZATION'
                trigger = 1
            else:
                if j[1] == NE:
                    NE_text += " " + text
                else:
                    clean.append((NE_text, NE))
                    NE_text = text
                    NE = j[1]
        elif j[1] == 'MONEY':
            if trigger == 0:
                NE_text = text
                NE = 'Money'
                trigger = 1
            else:
                if j[1] == NE:
                    NE_text += " " + text
                else:
                    clean.append((NE_text, NE))
                    NE_text = text
                    NE = j[1]
        elif j[1] == 'DATE':
            if trigger == 0:
                NE_text = text
                NE = 'DATE'
                trigger = 1
            else:
                if j[1] == NE:
                    NE_text += " " + text
                else:
                    clean.append((NE_text, NE))
                    NE_text = text
                    NE = j[1]
        elif j[1] == 'TIME':
            if trigger == 0:
                NE_text = text
                NE = 'TIME'
                trigger = 1
            else:
                if j[1] == NE:
                    NE_text += " " + text
                else:
                    clean.append((NE_text, NE))
                    NE_text = text
                    NE = j[1]

    if NE and NE_text:
        clean.append((NE_text, NE))

    clean_ = []
    for j in clean:
        text = re.sub(r'[^\w\s\u0E00-\u0E7F]','',j[0])
        text = lemmatizer.lemmatize(text.lower())
        if (text == '' or text == ' '):
            continue
        clean_.append(j)

    clean_token_eng.append(clean_)
```

# NP Chunking

```python
import nltk
```

```python
def pre_process(sentence, grammar):
    chunkParser = nltk.RegexpParser(grammar)
    tagged = nltk.pos_tag(nltk.word_tokenize(sentence))
    return chunkParser, tagged
```

```python
def get_chunks(tree):
    chunks = []
    for i in range(len(tree)):
        chunks.append(tree[i][:])
    return chunks
```

```python
with open('./data/thai_ner.data', 'rb') as filehandle:
    # read the data as binary data stream
    tmp = pickle.load(filehandle)
```

```python
clean_token_thai = []
for i in tqdm(tmp):
    clean = []
    NE_text = ''
    NE = ''
    trigger = 0
    for j in i:
        text = re.sub(r'[^\w\s\u0E00-\u0E7F]','',j[0])
        if (text == '' or text == ' ') and j[2][0] == 'O':
            continue
        
        if j[2][0] == 'B' and trigger == 0:
            NE_text = text
            NE = j[2][2:]
            trigger = 1
        elif j[2][0] == 'I' and trigger == 1:
            NE_text += text
        elif j[2][0] == 'B' and trigger == 1:
            clean.append((NE_text, NE))
            NE_text = text
            NE = j[2][2:]
        else:
            if NE_text and NE:
                clean.append((NE_text, NE))
                NE_text = ''
                NE = ''
                trigger = 0
            clean.append((text, j[2]))    
    if NE_text and NE:
        clean.append((NE_text, NE))
    clean_token_thai.append(clean)
```

```python
clean_token_eng = []
for i in tqdm(df['eng_tokenized_sents']):
    tmp = []
    for j in i:
        j = re.sub(r'[^\w\s\u0E00-\u0E7F]','',j)
        j = lemmatizer.lemmatize(j.lower())
        if j == '' or j == ' ':
            continue
        elif j in ['a', 'an', 'the']:
            continue
        tmp.append(j)
    clean_token_eng.append(tmp)
```

```python
df['clean_eng_tokenized'] = clean_token_eng
```
