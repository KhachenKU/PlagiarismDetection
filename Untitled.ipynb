```python
import matplotlib.pyplot as plt
```

```python
from tqdm import tqdm
```

```python
thai_side = []
eng_side = []
```

```python
with open("forward.align", 'r', encoding = "utf-16") as file:
    while True:
        line = file.readline().split()
        if not line:
            break
        line = [i.split('-') for i in line]
        thai = [int(i[0]) for i in line]
        eng = [int(i[1]) for i in line]
        thai_side.append(thai)
        eng_side.append(eng)
```

```python
all_diff_thai = []
all_diff_eng = []
```

```python
for _ in tqdm(range(len(eng_side))): 
    max_eng = 0
    max_thai = 0
    diff_thai = 0
    diff_eng = 0
    for i in range(len(eng_side[_])):
        if thai_side[_][i] <= max_thai:
            if diff_thai < max_thai-thai_side[_][i]:
                diff_thai = max_thai-thai_side[_][i]
                if diff_eng < eng_side[_][i]-max_eng:
                    diff_eng = eng_side[_][i]-max_eng
                continue

        max_eng = eng_side[_][i]
        max_thai = thai_side[_][i]
    all_diff_thai.append(diff_thai)
    all_diff_eng.append(diff_eng)
```

```python
dic_diff_eng = {}
dic_diff_thai = {}
```

```python
for eng, thai in zip(all_diff_thai,all_diff_eng):
    if eng not in dic_diff_eng:
        dic_diff_eng[eng] = 1
    else:
        dic_diff_eng[eng] += 1
        
    if thai not in dic_diff_thai:
        dic_diff_thai[thai] = 1
    else:
        dic_diff_thai[thai] += 1
```

```python
sorted_dic_diff_eng = sorted(dic_diff_eng.items())
```

```python
plt.plot([i[0] for i in sorted_dic_diff_eng[:20]], [i[1] for i in sorted_dic_diff_eng[:20]])
plt.xlabel("Length of the cross word")
plt.ylabel("Frequency")
plt.title("English")
```

```python
sorted_dic_diff_thai = sorted(dic_diff_thai.items())
```

```python
plt.plot([i[0] for i in sorted_dic_diff_thai[:20]], [i[1] for i in sorted_dic_diff_thai[:20]])
plt.xlabel("Length of the cross word")
plt.ylabel("Frequency")
plt.title("Thai")
```
